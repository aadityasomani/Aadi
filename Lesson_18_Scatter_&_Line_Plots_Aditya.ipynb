#### Slicing a DataFrame Using the `iloc[]` Function
Create Pandas series for the first 3 stars and the last 3 stars in the DataFrame:

````
# Create a Pandas series for the first star and store it in a variable called 'star_0'.
star_0 = exo_train_df.iloc[0, :]
star_0.head()
````

**Syntax:** 

`dataframe_name.iloc[row_position_start : row_position_end, column_position_start : column_position_end]`

In this syntax: 

- `row_position_start`: Denotes the position of the row in the DataFrame **starting** from whose values you want to take in the new Pandas series or DataFrame.
- `row_position_end`: Denotes the position of the row in the DataFrame till whose values you want to take in the new Pandas series or DataFrame.
- `column_position_start`: Denotes the position of the column in the DataFrame **starting** from whose values you want to take in the new Pandas series or DataFrame.
- `column_position_end`: Denotes the position of the column in the DataFrame till whose values you want to take in the new Pandas series or DataFrame.

You can verify manually whether we have extracted the values from the first row or not by viewing the first 5 rows of the DataFrame using the `head()` function:

````
# Create a Pandas series for the second star and store it in a variable called 'star_1'.
star_1 = exo_train_df.iloc[1, :]
star_1.head()
````

#### Task 1: Scatter and Line Plots for First 3 Stars

Let's plot the scatter plots and line plots for the first three planets in the `exo_train_df` DataFrame.

To make this plot, 

1. Import a Python module named `matplotlib.pyplot` with `plt` as an alias. This module is exclusively designed for creating graphs such as bar graphs, histograms, line plots, scatter plots, etc. We will learn more about this module as we go on in this course:

  ```
  import matplotlib.pyplot as plt
  ```

2. Call the `figure()` function from the `plt` module to resize the plot. The `figure()` function takes `figsize=(horizontal_width, vertical_height)` parameter as an input:  

  ```
  plt.figure(figsize=(16, 4))
  ```

3. Create either a Python list, a NumPy array or a Pandas series containing the numbers between `1` and `3197` to plot them on the $x-axis$:

  ```
  x_values_star_0 = np.arange(1, 3198)
  ```

4. Create the `star_0` Pandas series to plot the `FLUX` values on the $y-axis$ for the first star in the DataFrame: 

  ```
  y_values_star_0 = star_0[1:]
  ```

5. Call the `scatter()` function from the `plt` module with the required inputs as described in the third and the fourth steps:

  ```
  plt.scatter(x_values_star_0, y_values_star_0)
  ```

6. Call the `show()` function from the `plt` module:

  ```
  plt.show()
  ```

````
# S1.1: Create a scatter plot for the 'star_0' Pandas series.
# 1. Import the 'numpy' and 'matplotlib.pyplot' modules.
import matplotlib.pyplot as plt
import numpy as np
# 2. Call the 'figure()' function to resize the plot.
plt.figure(figsize=(24,10))
# Here, 16 means the plot is 16 units wide and 4 units high. Play with these numbers to draw different-sized plots.
x_values_star0=np.arange(1,3198)
y_values_star0=star_0[1:]
# Call the 'scatter()' function to make a scatter plot between the x and y values.
# The scatter() function requires two inputs: x and y where x is the data to be plotted on the x-axis and y is the data to be plotted on the y-axis.
# In our case, x is a Pandas series of numbers between 1 and 3197 and y is the 'FLUX' values for a star.
plt.scatter(x_values_star0,y_values_star0)
plt.show()


# Here, star_0[1:] is a Pandas series containing all the 'FLUX' values starting from the value at index 1 till the value at last index, i.e., 3197
# The 'arange(1, 3198)' function from the 'numpy' module will generate numbers from 1 to 3197.

# 3. Call the 'scatter()' function.

# 4. Call the 'show()' function.

# The 'show()' function displays the plot.
````

As we can see from the graph, there is a somewhat periodic downward peak in the brightness level recorded by the Kepler telescope for this star. This suggests that the first star in the DataFrame has at least one planet.

Let's draw the line plot also to see this fluctuation differently. 

To draw a line plot, you have to follow the same steps that you followed to create a scatter plot except in the fifth step, you have to call the `plot()` function instead of the `scatter()` function from the `plt` module:

````
# S1.2: Create a line plot for 'star_0' Pandas series.
# Line plot for the first star in the DataFrame.
plt.figure(figsize=(16,4))
plt.plot(x_values_star0,y_values_star0)
plt.show()
# Call the plot(x, y) function to draw a line plot between the x and y values.
````

The line plot also confirms the periodic downward peaks in the `FLUX` values.

````
# S1.3: Create a scatter plot for the second star, i.e., 'star_1'.
x_values_star1=np.arange(1,3198)
y_values_star1=star_1[1:]
plt.figure(figsize=(16,4))
plt.scatter(x_values_star1,y_values_star1)
plt.show()
````

In the above code, we plotted a scatter plot for the first star. However, it is quite difficult to spot any clear pattern in the scatters plot for the second star in the DataFrame. 

Let's draw a line plot to identify a pattern:

````
# S1.4: Create a line plot for the second star, i.e., 'star_1'.
plt.figure(figsize=(16,4))
plt.plot(x_values_star1,y_values_star1)
plt.show()
````

As we can see, there are consistent sudden drops in the brightness levels for the second star in the DataFrame. This suggests that the planet is orbiting its star at a very high radial speed. Also, the planet could be very close to the star.



````
# S1.5: Create a scatter plot for the third star, i.e., 'star_2'.
x_values_star2=np.arange(1,3198)
y_values_star2=star_2[1:]
plt.figure(figsize=(16,4))
plt.scatter(x_values_star2,y_values_star2)
plt.show()
````

Here also, we can spot clear repetitive downward-peaks which confirms that the star has at least one planet.

````
# S1.5: Create a line plot for the third star, i.e, 'star_2'.
plt.figure(figsize=(16,4))
plt.plot(x_values_star2,y_values_star2)
plt.show()
````

The line plot also confirms the repetitive downward-peak pattern.

#### Task 2: Scatter Plots and Line Plots for Last 3 Stars
Now, let's create the scatter plots and line plots for the last three stars in the DataFrame which have been labeled or classified as `1`:

````
# S2.1: Create a scatter plot for the last star, i.e., 'star_5086' in the DataFrame.
x_values_star5086=np.arange(1,3198)
y_values_star5086=star_5086[1:]
plt.figure(figsize=(16,4))
plt.scatter(x_values_star5086,y_values_star5086)
plt.show()
````

As you can see, there is no clear periodic downward peak in the brightness levels of the light emitted from the star. We cannot claim that the star has a planet orbiting it. Let's also create the line plot to confirm our inference:

````
# S2.2: Create a line plot for the last star, i.e, 'star_5086' in the DataFrame.
plt.figure(figsize=(16,4))
plt.plot(x_values_star5086,y_values_star5086)
plt.show()
````

There are some sudden spikes in the line plot but they exist at irregular intervals. So we cannot say for sure that the star has a planet.

Let us create a scatter plot for the second-last star:

````
# S2.3: Create a scatter plot for the second-last star, i.e., 'star_5085' in the DataFrame.
x_values_star5085=np.arange(1,3198)
y_values_star5085=star_5085[1:]
plt.figure(figsize=(16,4))
plt.scatter(x_values_star5085,y_values_star5085)
plt.show()
````

There is no clear periodic downward-peak pattern in the `FLUX` values for the second-last star.

````
# S2.4: Create a line plot for the second-last star in the DataFrame.
plt.figure(figsize=(16,4))
plt.plot(x_values_star5085,y_values_star5085)
plt.xlabel("Time")
plt.ylabel("Brightness")
plt.show()
````

The line-plot also confirms that there is no clear periodic downward-peak pattern in the `FLUX` values.

````
# S2.5: Create a scatter plot for the third-last star, i.e., 'star_5084' in the DataFrame.
x_values_star5084=np.arange(1,3198)
y_values_star5084=star_5084[1:]
plt.figure(figsize=(16,4))
plt.scatter(x_values_star5084,y_values_star5084,label="No planet")
plt.scatter(x_values_star1,y_values_star1,label="Has planet")
plt.legend()
plt.xlabel("Time")
plt.ylabel("Brightness")
plt.show()
````

No clear periodic downward-peak pattern for the third-last star.
